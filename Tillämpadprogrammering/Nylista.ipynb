```python
def multi(n):
    print('x', end='\t')
    print('\t'.join(str(i) for i in range(1, n+1)))
    for r in range(1, n+1):
        print(r, end='\t')
        print('\t'.join(str(r*c) for c in range(1, n+1)))

n = int(input("Välj ett tal för att visa dess multiplikationstabell."))

multi(n)
```
