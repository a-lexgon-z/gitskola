```python
# Uppgift 1 C

import pandas as pd
import matplotlib.pyplot as plt

chart = []

df = pd.read_csv("sysselsattning.csv", encoding="ISO-8859-1", header=1)
affermative = df["2019"].iloc[0:12].sum()
negative = df["2019"].iloc[12:24].sum()

chart.append(affermative)
chart.append(negative)
titlar = ["Är förvärvsarbetande", "Är inte förvärvsarbetande"]
color = ["gold", "silver"]

plt.pie(chart, labels=titlar, colors=color, autopct="%.2f%%", shadow=True)
plt.title("Andel som är förvärvsarbetande och som inte är")
plt.show()
```

```python
# Uppgift 2 E

import pandas as pd
import matplotlib.pyplot as plt

pog = []
thanos = "Förvärvsarbetande \nmän", "Förvärvsarbetande \nkvinnor", "Ej förvärvsarbetande \nmän", "Ej förvärvsarbetande \nkvinnor"

df = pd.read_csv("sysselsattning.csv", header=1)

manpog = df.query("sysselsättning == 'förvärvsarbetande' and kön == 'män'")["2019"].sum()
mannotpog = df.query("sysselsättning == 'förvärvsarbetande' and kön == 'kvinnor'")["2019"].sum()
kvinnorpog = df.query("sysselsättning == 'ej förvärvsarbetande' and kön == 'män'")["2019"].sum()
kvinnornotpog = df.query("sysselsättning == 'ej förvärvsarbetande' and kön == 'kvinnor'")["2019"].sum()

pog.append(manpog)
pog.append(mannotpog)
pog.append(kvinnorpog)
pog.append(kvinnornotpog)

font = {'size': 8}
plt.rc('font', **font)
plt.title("Sysselsättning för män och kvinnor mellan åldrarna 16-44")
plt.bar(thanos, pog, width=.2, color=["black", "yellow", "black", "yellow"])
plt.show()
```
