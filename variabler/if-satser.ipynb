```python
tal = int(input("Ange ett heltal: "))

if tal >0:
    print(f"{tal} är ett positivt tal")
elif tal == 0:
    print("Noll är varken negativt eller positivt")
else:
    print(f"{tal} är ett negativt tal")
```

```python
tal = int(input("Ange ett heltal: "))

if tal % 2:
    print(f"{tal} är ett udda tal")
else:
    print(f"{tal} är ett jämnt tal")

if tal % 5 == 0:
    print("Ditt tal är delbart med 5")
```

```python
tal1 = float(input("Ange det första talet: "))
tal2 = float(input("Ange det andra talet: "))

if tal1 > tal2:
    print(f"{tal1} > {tal2}")
elif tal2 > tal1:
    print(f"{tal1} < {tal2}")
```

```python
vinkel = float(input("Ange en grad vinkel: "))

if vinkel == 0:
    print("Din vikel är 0°")
elif vinkel == 90:
    print("Rät vinkel")
elif vinkel == 180:
    print("Halv vinkel")
elif vinkel == 360:
    print("Hel vinkel")
elif vinkel > 90 < 180:
    print("Trubbig vinkel")
elif vinkel > 0 < 90:
    print("Spetsig vinkel")
else:
    print("Konvex vinkel")
```
